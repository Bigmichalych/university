```python
import random

x = random.randint(0,100)
print("Компьютер загадал число: ", x)
k = 0
while 1 == 1:
  y = input("Ввод пользователя ")
  if y == 'Выход':
    k = k + 1
    break
  elif int(y) > x:
    k = k + 1
    print("Ваше число больше ")
    k = k + 1
  elif int(y) < x:
    k = k + 1
    print("Ваше число меньше ")
  elif int(y) == x:
    break
if k == 0:
  k = 1
if y == 'Выход':
  pass
elif int(y) == x:
  print("Поздоавляем! Вы угадали число за", k, "попытки!")



```
